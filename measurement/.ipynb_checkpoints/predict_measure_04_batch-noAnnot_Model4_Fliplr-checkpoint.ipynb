```python
image_ids = []

for file in os.listdir(dataset_dir)[1:]: # exclude the json
    image_path = image_path = os.path.join(dataset_dir, file)
    image_ids.append(image_path)
    
image_ids
```

```python
#Loading the model in the inference mode
config = pineapple.PineappleConfig()
model = modellib.MaskRCNN(mode="inference", config=config, model_dir='./')

# loading the trained weights on the custom dataset
model_path = "C:/Users/jessi/Mask_RCNN/logs/pineapple20201116T0825/mask_rcnn_pineapple_0027.h5"
model.load_weights(model_path, by_name=True)


image_ids = image_ids #from cell above
fruitsize = []

for image_id in image_ids:
    # Load image
    # load image and convert to array
    img = load_img(image_id)
    imgArray = img_to_array(img)
    
    # Run object detection
    results = model.detect([imgArray], verbose=0)
    
    # save predicted mask output
    pred_masks = results[0]['masks']
    pred_masks = np.transpose(pred_masks, [2, 0, 1])
    
    # Extract sizes
    
    for mask in pred_masks:
    
        binarymask = mask.astype(int)
    
        # FindContours support only 32s and 8u
        binarymask = binarymask.astype(np.uint8)

        # three arguments in cv.findContours() function, 
        # first one is source image, 
        # second is contour retrieval mode, 
        # third is contour approximation method. 
        # And it outputs a modified image, the contours and hierarchy.

        cnts = cv2.findContours(binarymask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[0]
    
        # initialise somewhere to store the coord points from each contour
        pts = np.empty((len((cnts)[0]), 2))
        # then loop through and get coords into tuples in array
        for i in range(len(cnts[0])):
            pts[i] = ((cnts)[0][i][0][0], (cnts)[0][i][0][1])
        # make sure pts contains integers not floats
        pts = np.int32(pts)
    
        # need to reshape array 
        pts = pts.reshape((-1,1,2))
    
        box = cv2.minAreaRect(pts)
        (x, y), (width, height), angle = box
    
        unsorted_tuple = (round(width,2), round(height, 2))
        
        fruitsize.append(tuple(sorted(unsorted_tuple)))
    
        print(round(width,2), round(height, 2))
    
```
